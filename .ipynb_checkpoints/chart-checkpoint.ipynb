```python
def get_repurchase_raw_data(day): 
    start = datetime.datetime.now() - datetime.timedelta(day)
    end = datetime.datetime.now()
    dates = pd.bdate_range(start, end)
    column_name = ['Stock', 'Code', 'Sec_type', 'Date', 'Num_sec', 'Highest_price', 'Lowest_price', 'Total_paid', 'Method', 'Number_of_sec_purchased', 'Perc_number_of_shares_issue']
    output_df = pd.DataFrame()

    for date in dates:
        try:
            # Extract excel file from hkex
            end = str(date).split()[0].replace('-','')
            temp_df = pd.read_excel("repurchase_data/{}.xls".format(end))

            # Get the index that matches my condition
            temp_df = temp_df.loc[temp_df['Unnamed: 1'] < '9999999']
            output_df = output_df.append(temp_df)
            output.close()
        except:
            pass

    # Change name
    output_df = output_df.rename(columns={'Unnamed: 0': 'Stock', 'Unnamed: 1': 'Code', 'Unnamed: 2': 'Sec_type', 'Unnamed: 3': 'Date', 
                    'Unnamed: 4': 'Num_sec', 'Unnamed: 5': 'Highest_price', 'Unnamed: 6': 'Lowest_price', 'Unnamed: 7': 'Total_paid', 
                    'Unnamed: 8': 'Method', 'Unnamed: 9' :'Number_of_sec_purchased', 'Unnamed: 10' :'Perc_number_of_shares_issue'})

    return output_df, dates
```

```python
clean_repurchase_df = repurchase_df[0]
for index, row in clean_repurchase_df.iterrows():
    try:
        row['Num_sec'] = int(str(row['Num_sec']).replace(',',''))
        row['Total_paid'] = int(str(row['Total_paid']).split(" ")[1].split(".")[0].replace(',',''))
        row['Date'] = row['Date'].split()[0].replace('/','-')
    except :
        row['Total_paid'] = 0
clean_repurchase_df
```

```python
clean_repurchase_df['Date']= pd.to_datetime(clean_repurchase_df['Date'])
 
# Check the format of 'Date' column
clean_repurchase_df.info()
```

```python
hsi_df['Date']= pd.to_datetime(hsi_df['Date'])
hsi_df.info()
```

```python
total_repurchase_date_df = clean_repurchase_df.groupby('Date').agg({'Total_paid': 'sum'}).reset_index()
```

```python
total_repurchase_date_df
```

```python
total_num_date_df = clean_repurchase_df.groupby(['Date']).size().to_frame('Num comp')
total_num_date_df
```

```python
total_repurchase_date_df.index = total_repurchase_date_df['Date']
total_repurchase_date_df
```

```python
new_df = total_num_date_df.merge(total_repurchase_date_df, left_index=True, right_index=True, how='inner')
new_df
```

```python
news_df = new_df['20170101':'20211230']
news_df
```
